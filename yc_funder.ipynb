```python
# Combine  pickle files
folder_path = "data"
soup_list = []

for filename in os.listdir(folder_path):
    if filename.endswith(".pkl"):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'rb') as file:
            soup = pickle.load(file)
            soup_list.append(soup)

combined_html = ""
for soup in soup_list:
    combined_html += str(soup)
```

```python
soup = BeautifulSoup(combined_html, "html.parser")
```

```python
# Extract all links

links = soup.find_all("a", attrs={'class':'_company_99gj3_339'})
len(links)
```

```python
links_list = []
for link in links:
    links_list.append(link.get('href'))
 
```

```python

links_list
```

```python
# Define the function to get founders details

def get_founders_details(soup, company_name):
    founders = []

    founders_data = soup.find_all("div", class_="flex flex-row flex-col items-start gap-3 md:flex-row")

    for founder_data in founders_data:
        founder = {}

        # Name
        founder['Name'] = founder_data.find("h3", class_='text-lg font-bold').text.strip()

        # Role
        founder['Role'] = founder_data.find("h3", class_='text-lg font-bold').text.strip()

        # Description
        founder['Biography'] = founder_data.find("p", class_='prose max-w-full whitespace-pre-line').text.strip()

        # LinkedIn
        linkedin_link_tag = founder_data.find("a", class_='bg-image-linkedin')
        founder['LinkedIn'] = linkedin_link_tag['href'] if linkedin_link_tag else None

        # Twitter
        twitter_link_tag = founder_data.find("a", class_='bg-image-twitter')
        founder['Twitter'] = twitter_link_tag['href'] if twitter_link_tag else None

        # Company Name
        founder['Company'] = company_name

        founders.append(founder)

    return founders
```

```python
# Define dictionary to store founders data
d = {
    "Company": [],  
    "Name": [],
    "Role": [],
    "Biography": [],
    "LinkedIn": [],
    "Twitter": []
}

i=4492

for link in links_list:
    url = "https://www.ycombinator.com" + link
    response = requests.get(url)
    
    if response.status_code == 200:
        webpage_content = response.content
        new_soup = BeautifulSoup(webpage_content, "html.parser")
        

        company_name_tag = new_soup.find("h1", class_='font-extralight')
        company_name = company_name_tag.text.strip() if company_name_tag else None

        founders = get_founders_details(new_soup, company_name)
        
        for founder in founders:
            d['Company'].append(founder['Company'])  
            d['Name'].append(founder['Name'])
            d['Role'].append(founder['Role'])
            d['Biography'].append(founder['Biography'])
            d['LinkedIn'].append(founder['LinkedIn'])
            d['Twitter'].append(founder['Twitter'])
            print(i, "left..")
            i=i-1
    else:
        print(f"Failed to retrieve data from {url}")


```

```python
# Converting to data frame

df = pd.DataFrame(d)


df.to_csv(' founder.csv', index=False)
```

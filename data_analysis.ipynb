```python
# Code for create Company funder details has a json file

company_founder_list = []  

company_founders = {}


for _, row in df2.iterrows():
    company_name = row['Company']
    founder_dict = {
        "name": row['Name'],
        "role": row['Role'],
        "description": row['Biography'],
        "twitter_profile": row['Twitter'],
        "linkedin_profile": row['LinkedIn']
    }
    if company_name not in company_founders:
        company_founders[company_name] = []
    company_founders[company_name].append(founder_dict)


for _, row in df1.iterrows():
    company_dict = {
        "company": {
            "name": row['Name'],
            "tagline": row['Tagline'],
            "description": row['Description'],
            "batch": row['Batch'],
            "company_type": row['Companytype'],
            "industry_tags": row['Industrytags'].split(', ') if isinstance(row['Industrytags'], str) else [],
            "location": row['Location'],
            "website": row['Website'],
            "founded": row['Yearfounded'],
            "team_size": row['Teamsize'],
            "social_profiles": {
                "linkedin": row['linkedin'],
                "twitter": row['Twitter'],
                "facebook": row['Facebook'],
                "crunchbase": row['Crunchbase']
            }
        },
        "founders": company_founders.get(row['Name'], []) 
    }
    company_founder_list.append(company_dict)

with open('output/output.json', 'w') as json_file:
    json.dump(company_founder_list, json_file, indent=4)
```
